```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load clean data from part 1
df_cr = 'data/Clean_Mortgage_Lending.csv'
df_cr = pd.read_csv(df_cr, sep = ',')
```

```python
df_cr.info()
```

```python
df_cr.nunique()
```

```python
df_cr['action_taken'].mask((df_cr['action_taken'] == 1), 'Approved', inplace=True)
df_cr['action_taken'].mask((df_cr['action_taken'] == 0), 'Denied', inplace=True)
```

```python
plt.rcParams['figure.dpi'] = 500
sns.set(rc={"figure.dpi":500, 'savefig.dpi':500})
['#ff9999','#99ccff','#99ffcc','#ffcc99', '#9999ff', '#ffff99'] # red, blue, green, orange, purple, yellow
```

```python
# a function to explore each different category
def singlePie(column_name, colors, viz_title):
    v = df_cr[column_name].value_counts().plot(kind='pie', colors = colors, autopct='%1.1f%%', labeldistance = 1.05)
    v.yaxis.set_visible(False)
    plt.axis('equal')
    plt.tight_layout()
    centre_circle = plt.Circle((0,0),0.75,fc='white')
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)
    #plt.rcParams['font.size'] = 9.0
    plt.title(viz_title, fontsize=15, fontweight=2, pad = 10)
```

```python
singlePie('action_taken', ['#99ffcc', '#ff9999'], 'Action Taken')
```

```python
# a function to explore each different category
def doublePie(outer_column_name, outer_colors, viz_title, inner_column_name, inner_colors):
    v = df_cr[outer_column_name].value_counts().plot(kind='pie', colors = outer_colors, autopct='%1.1f%%', labeldistance = 1.05, frame = True)
    v.yaxis.set_visible(False)
    plt.axis('equal')
    plt.tight_layout()
    #plt.rcParams['font.size'] = 9.0
    plt.title(viz_title, fontsize=15, fontweight=2, pad = 10)
    plt.pie(inner_column_name,colors=inner_colors,radius=0.75)
    centre_circle = plt.Circle((0,0),0.5, color='black', fc='white', linewidth=0)
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)
    
```

### Bias + Income

```python
# extract them with the lowest debt to income ratio
rich = rich.loc[(rich['debt_to_income_ratio'] == '<20%') | (rich['debt_to_income_ratio'] == '20%-<30%')]
```
